```python
class PersonalData:
    def __init__(self, name, address, age, phone_number):
        # Constructor method to initialize the personal data attributes
        self.name = name  # Assign the name attribute
        self.address = address  # Assign the address attribute
        self.age = age  # Assign the age attribute
        self.phone_number = phone_number  # Assign the phone number attribute

    # Accessor methods
    def get_name(self):
        # Method to get the value of the name attribute
        return self.name

    def get_address(self):
        # Method to get the value of the address attribute
        return self.address

    def get_age(self):
        # Method to get the value of the age attribute
        return self.age

    def get_phone_number(self):
        # Method to get the value of the phone number attribute
        return self.phone_number

    # Mutator methods
    def set_name(self, name):
        # Method to set the value of the name attribute
        self.name = name

    def set_address(self, address):
        # Method to set the value of the address attribute
        self.address = address

    def set_age(self, age):
        # Method to set the value of the age attribute
        self.age = age

    def set_phone_number(self, phone_number):
        # Method to set the value of the phone number attribute
        self.phone_number = phone_number

def main():
    # Create instances of PersonalData class
    my_info = PersonalData("Your Name", "Your Address", 25, "Your Phone Number")
    friend1_info = PersonalData("Friend 1's Name", "Friend 1's Address", 30, "Friend 1's Phone Number")
    friend2_info = PersonalData("Friend 2's Name", "Friend 2's Address", 35, "Friend 2's Phone Number")

    # Display personal data
    print("Your Information:")
    print("Name:", my_info.get_name())
    print("Address:", my_info.get_address())
    print("Age:", my_info.get_age())
    print("Phone Number:", my_info.get_phone_number())

    print("\nFriend 1's Information:")
    print("Name:", friend1_info.get_name())
    print("Address:", friend1_info.get_address())
    print("Age:", friend1_info.get_age())
    print("Phone Number:", friend1_info.get_phone_number())

    print("\nFriend 2's Information:")
    print("Name:", friend2_info.get_name())
    print("Address:", friend2_info.get_address())
    print("Age:", friend2_info.get_age())
    print("Phone Number:", friend2_info.get_phone_number())

if __name__ == "__main__":
    main()
```

```python
class TriviaQuestion:
    def __init__(self, question, options, correct_answer):
        # Constructor method to initialize the trivia question attributes
        self.question = question  # Assign the question attribute
        self.options = options  # Assign the options attribute
        self.correct_answer = correct_answer  # Assign the correct_answer attribute

    # Accessor methods
    def get_question(self):
        # Method to get the value of the question attribute
        return self.question

    def get_options(self):
        # Method to get the value of the options attribute
        return self.options

    def get_correct_answer(self):
        # Method to get the value of the correct_answer attribute
        return self.correct_answer

def main():
    # Create a list of trivia questions
    trivia_questions = [
        TriviaQuestion("What is the capital of France?", ["1. Paris", "2. Rome", "3. Berlin", "4. Madrid"], 1),
        TriviaQuestion("Who painted the Mona Lisa?", ["1. Leonardo da Vinci", "2. Pablo Picasso", "3. Vincent van Gogh", "4. Michelangelo"], 1),
        TriviaQuestion("Which planet is known as the Red Planet?", ["1. Venus", "2. Mars", "3. Jupiter", "4. Saturn"], 2),
        TriviaQuestion("Who wrote 'Romeo and Juliet'?", ["1. William Shakespeare", "2. Charles Dickens", "3. Jane Austen", "4. Mark Twain"], 1),
        TriviaQuestion("What is the tallest mammal?", ["1. Elephant", "2. Giraffe", "3. Hippopotamus", "4. Rhino"], 2),
        TriviaQuestion("What is the chemical symbol for water?", ["1. H2O", "2. CO2", "3. NaCl", "4. O2"], 1),
        TriviaQuestion("Which country is famous for its tulips?", ["1. Italy", "2. France", "3. Netherlands", "4. Germany"], 3),
        TriviaQuestion("Who discovered penicillin?", ["1. Isaac Newton", "2. Marie Curie", "3. Alexander Fleming", "4. Albert Einstein"], 3),
        TriviaQuestion("What is the largest ocean on Earth?", ["1. Atlantic Ocean", "2. Indian Ocean", "3. Pacific Ocean", "4. Arctic Ocean"], 3),
        TriviaQuestion("Who invented the telephone?", ["1. Thomas Edison", "2. Alexander Graham Bell", "3. Nikola Tesla", "4. James Watt"], 2)
    ]

    # Initialize players' scores
    player1_score = 0
    player2_score = 0

    # Iterate through each trivia question
    for i, question in enumerate(trivia_questions, 1):
        print(f"\nQuestion {i}:")
        print(question.get_question())  # Display the question
        for option in question.get_options():
            print(option)  # Display each option
        answer = int(input("Enter your answer (1-4): "))  # Prompt the player for an answer

        # Check if the answer is correct
        if answer == question.get_correct_answer():
            print("Correct!")
            # Assign points to the respective player based on the question number
            if i % 2 == 1:
                player1_score += 1
            else:
                player2_score += 1
        else:
            print("Incorrect!")

    # Display players' scores and declare the winner
    print("\nPlayer 1 Score:", player1_score)
    print("Player 2 Score:", player2_score)
    if player1_score > player2_score:
        print("Player 1 wins!")
    elif player2_score > player1_score:
        print("Player 2 wins!")
    else:
        print("It's a tie!")

if __name__ == "__main__":
    main()
```
